# Model Evaluation

Goal:
Verify whether the classifier is truly learning.

Questions:

- Is accuracy reliable?
- Are classes balanced?
- Does the model generalize?

```python
from pathlib import Path

import joblib
import numpy as np
```

```python
ROOT = Path("../")

X = np.load(ROOT / "data/processed/X.npy")

y = np.load(ROOT / "data/processed/y.npy")

model = joblib.load(ROOT / "models/baseline.pkl")
```

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(model, X, y, cv=5)

scores
```

```python
print("mean:", scores.mean())
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

pred = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(8, 8))

plt.imshow(cm)

plt.show()
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, pred))
```

```python
probs = model.predict_proba(X_test)
```

```python
probs.max(axis=1).mean()
```
