# Audio Representation

Goal:
Understand how audio becomes model input.

Questions:

- What is waveform?
- What is spectrogram?
- What is mel spectrogram?
- Can different sounds be visually different?

```python
from pathlib import Path

import librosa
import librosa.display

import matplotlib.pyplot as plt

import numpy as np
```

```python
ROOT = Path("../data/raw/esc50")

AUDIO = ROOT / "audio"

sample = AUDIO / "1-100032-A-0.wav"
```

```python
y, sr = librosa.load(
    sample,
    sr=None
)

print("sample rate:", sr)

print("shape:", y.shape)
```

## Waveform

```python
plt.figure(figsize=(12,4))

librosa.display.waveshow(
    y,
    sr=sr
)

plt.show()
```

## Spectrogram

```python
S = np.abs(
    librosa.stft(y)
)
```

```python
plt.figure(figsize=(12,5))

librosa.display.specshow(
    librosa.amplitude_to_db(
        S
    ),
    sr=sr,
    x_axis="time",
    y_axis="hz"
)

plt.colorbar()

plt.show()
```

## Mel Spectrogram

```python
mel = librosa.feature.melspectrogram(
    y=y,
    sr=sr
)
```

```python
plt.figure(figsize=(12,5))

librosa.display.specshow(
    librosa.power_to_db(
        mel
    ),
    x_axis="time",
    y_axis="mel",
    sr=sr 
)

plt.colorbar()

plt.show()
```

# Findings

Waveform:
Hard to interpret.

Spectrogram:
Shows energy.

Mel:
Much easier to distinguish sounds.
