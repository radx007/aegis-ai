# Dataset Exploration — Aegis V1

Goal:
Understand ESC50 and select sounds relevant for Aegis.

Dataset:
ESC50

Output:
Selected classes for baseline.

```python
from pathlib import Path
import pandas as pd
```

```python
ROOT = Path("../data/raw/esc50")

META = ROOT / "meta" / "esc50.csv"

AUDIO = ROOT / "audio"
```

```python
df = pd.read_csv(META)

df.head()
```

## Dataset Summary

```python
print("Total samples:", len(df))

print("Categories:")

print(df["category"].nunique())
```

```python
sorted(df["category"].unique())
```

## Aegis Target Sounds

```python
TARGET_CLASSES = {
    "alarm": ["clock_alarm"],
    "baby_cry": ["crying_baby"],
    "siren": ["siren"],
    "knock": ["door_wood_knock"]
}
```

## Keep Relevant Sounds

```python
selected = []

for values in TARGET_CLASSES.values():
    selected.extend(values)

subset = df[
    df["category"].isin(selected)
]

subset.head()

subset["category"].value_counts()
```

```python
from IPython.display import Audio
```

```python
sample = subset.iloc[3]

path = AUDIO / sample["filename"]

Audio(path)
```

```python
import librosa
```

```python
def duration(path):

    y, sr = librosa.load(path)

    return len(y) / sr
```

```python
duration(path)
```

```python
subset["duration"] = subset["filename"].apply(
    lambda x:
    duration(AUDIO / x)
)

subset["duration"].describe()
```
